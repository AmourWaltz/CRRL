```python
model = "qwen25_3b_it"
input_path = "./exp/{}_search/{}_results.jsonl"

datasets = ["nq", "triviaqa", "popqa", "hotpotqa", "2wikimultihopqa", "musique", "bamboogle"]

from search_r1.utils import *
import numpy as np
from verl.utils.reward_score import qa_em

recall_rate = []
accuracy = []
recall_accu = []

def _select_rm_score_fn(data_source):
    if data_source in ['nq', 'triviaqa', 'popqa', 'hotpotqa', '2wikimultihopqa', 'musique', 'bamboogle']:
        return qa_em.compute_score_em
    else:
        raise NotImplementedError


def test_evaluate(data_pool, dataset):
    total_scores = []
    total_search = []
    entail = 0
    false_err = 0
    for data in data_pool:
        # import pdb; pdb.set_trace()
        # select reward score based on data_source
        ground_truth = data["golden_answers"]
        generation = data["output"]["generation"]
        sequences_str = data["output"]["answer"]
        search_times = data["output"].get("search_times", 0)

        compute_score_fn = _select_rm_score_fn(dataset)
        

        score = compute_score_fn(solution_str=sequences_str, ground_truth=ground_truth, format_score=0)
        total_scores.append(score)
        total_search.append(search_times)
        
        for ground in ground_truth:
            if ground in generation:
                # print("Found in generation:", ground)
                entail += 1
                if score == 1:
                    false_err += 1
                break

    print(f"Dataset: {dataset}, Search times: {np.mean(total_search)}, Total: {len(total_search)}, Score: {np.mean(total_scores)}, Entail Rate: {entail/len(total_search)}, False Error: {false_err/len(total_search)}")
    recall_rate.append(entail/len(total_search))
    accuracy.append(np.mean(total_scores))
    recall_accu.append(false_err/len(total_search))

for dataset in datasets:
    data_pool = read_jsonl(input_path.format(model, dataset))
    test_evaluate(data_pool, dataset)
```

```python
from search_r1.utils import *

import transformers

datasets = ["nq", "hotpotqa"]

def token_lens(tokenizer, input_path):
    data_pool = read_jsonl(input_path)[:1000]
    lens = []
    for item in data_pool:
        context = item["question"] + "\n" + item["output"]["generation"]
        # Calculate the context length
        context_len = len(tokenizer(context, return_tensors="pt").input_ids[0])
        lens.append(context_len)
    return sum(lens) / len(lens)


models = ["qwen25_7b_it"]
# methods = ["search", "search_extract_v2"]
methods = ["search_selfref"]

for model in models:
    tokenizer = transformers.AutoTokenizer.from_pretrained(model_dict[model], trust_remote_code=True)
    for method in methods:
        lens = 0
        for dataset in datasets:
            input_path = "./exp/{}_{}/{}_results.jsonl".format(model, method, dataset)
            lens += token_lens(tokenizer, input_path)
        print(f"Model: {model}, Method: {method}, Overall Avg Token Len: {lens/len(datasets)}")
```

```python
model = "qwen25_3b_it"
input_path = "./exp/{}_search_think/{}_results.jsonl"

datasets = ["nq", "triviaqa", "popqa", "hotpotqa", "2wikimultihopqa", "musique", "bamboogle"]

from search_r1.utils import *
import numpy as np
from verl.utils.reward_score import qa_em

recall_rate = []
accuracy = []
recall_accu = []

def _select_rm_score_fn(data_source):
    if data_source in ['nq', 'triviaqa', 'popqa', 'hotpotqa', '2wikimultihopqa', 'musique', 'bamboogle']:
        return qa_em.compute_score_em
    else:
        raise NotImplementedError


def test_evaluate(data_pool, dataset):
    total_scores = []
    total_search = []
    entail = 0
    false_err = 0
    for data in data_pool:
        # import pdb; pdb.set_trace()
        # select reward score based on data_source
        ground_truth = data["golden_answers"]
        generation = data["output"]["generation"]
        sequences_str = data["output"]["answer"]
        search_times = data["output"].get("search_times", 0)

        compute_score_fn = _select_rm_score_fn(dataset)
        

        score = compute_score_fn(solution_str=sequences_str, ground_truth=ground_truth, format_score=0)
        total_scores.append(score)
        total_search.append(search_times)
        
        for ground in ground_truth:
            if ground in generation and "<information>" in generation:
                # print("Found in generation:", ground)
                entail += 1
                if score == 1:
                    false_err += 1
                break

    # print(f"Dataset: {dataset}, Search times: {np.mean(total_search)}, Total: {len(total_search)}, Score: {np.mean(total_scores)}, Entail Rate: {entail/len(total_search)}, False Error: {false_err/len(total_search)}")
    recall_rate.append(entail/len(total_search))
    accuracy.append(np.mean(total_scores))
    recall_accu.append(false_err/len(total_search))

for dataset in datasets:
    data_pool = read_jsonl(input_path.format(model, dataset))[:1000]
    test_evaluate(data_pool, dataset)

print(f"Overall Recall Rate: {round(np.mean(recall_rate), 3)}, Overall Accuracy: {round(np.mean(accuracy), 3)}, Overall Recall Accu: {round(np.mean(recall_accu), 3)}")
```
