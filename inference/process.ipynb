```python
from search_r1.utils import *
import numpy as np

from verl.utils.reward_score import qa_em

def _select_rm_score_fn(data_source):
    if data_source in ['nq', 'triviaqa', 'popqa', 'hotpotqa', '2wikimultihopqa', 'musique', 'bamboogle']:
        return qa_em.compute_score_em
    else:
        raise NotImplementedError

input_path_1 = "./exp/qwen25_3b_it_search/nq_results.jsonl"
input_path_2 = "./exp/qwen25_3b_it_search_lastklg/nq_results.jsonl"

data_pool_1 = read_jsonl(input_path_1)
data_pool_2 = read_jsonl(input_path_2)

items_1, items_2, cnt = [], [], 0
eval_data = []

for item1, item2 in zip(data_pool_1, data_pool_2):
    assert item1["id"] == item2["id"]
    if item1["output"]["search_times"] > 1 and item2["output"]["search_times"] > 1:
        cnt += 1
        items_1.append(item1)
        items_2.append(item2)


def test_evaluate(data_pool):
    total_scores = []
    total_search = []

    for data in data_pool:
        # import pdb; pdb.set_trace()
        # select reward score based on data_source
        ground_truth = data["golden_answers"]
        sequences_str = data["output"]["answer"]
        search_times = data["output"].get("search_times", 0)
        if search_times == 1:
            continue

        compute_score_fn = _select_rm_score_fn("nq")

        score = compute_score_fn(solution_str=sequences_str, ground_truth=ground_truth, format_score=0)
        total_scores.append(score)
        total_search.append(search_times)
    print(f"Score: {np.mean(total_scores)}, Search times: {np.mean(total_search)}, Total: {len(total_search)}")


print(len(items_1), len(items_2), cnt)
test_evaluate(items_1)
test_evaluate(items_2)
```

```python
from search_r1.utils import *

import transformers

datasets = ["nq", "hotpotqa"]

def token_lens(tokenizer, input_path):
    data_pool = read_jsonl(input_path)[:1000]
    lens = []
    for item in data_pool:
        context = item["question"] + "\n" + item["output"]["generation"]
        # Calculate the context length
        context_len = len(tokenizer(context, return_tensors="pt").input_ids[0])
        lens.append(context_len)
    return sum(lens) / len(lens)


models = ["qwen25_7b_it"]
# methods = ["search", "search_extract_v2"]
methods = ["search_selfref"]

for model in models:
    tokenizer = transformers.AutoTokenizer.from_pretrained(model_dict[model], trust_remote_code=True)
    for method in methods:
        lens = 0
        for dataset in datasets:
            input_path = "./exp/{}_{}/{}_results.jsonl".format(model, method, dataset)
            lens += token_lens(tokenizer, input_path)
        print(f"Model: {model}, Method: {method}, Overall Avg Token Len: {lens/len(datasets)}")
```

### Processing context refinement training data

```python
from utils import *

def get_query(text):
    import re
    pattern = re.compile(r"<search>(.*?)</search>", re.DOTALL)
    matches = pattern.findall(text)
    return matches

input_path_1 = "./../exp/qwen25_3b_it_search_extract/nq_results.jsonl"
data_pool_1 = read_jsonl(input_path_1)[1000:]
print(len(data_pool_1))

input_path_2 = "./../exp/qwen25_7b_it_search_extract/nq_results.jsonl"
data_pool_2 = read_jsonl(input_path_2)[1000:]
print(len(data_pool_2))

data_pool = data_pool_1 + data_pool_2
print(len(data_pool))
print(data_pool[0]["output"].keys())

idx = 0

new_data_pool = []
for item in data_pool:
    queries = get_query(item["output"]["generation"])
    for doc, refine, query in zip(item["output"]["search_docs"], item["output"]["extract_docs"], queries):
        new_data = {
            "id": idx,
            "context": doc,
            "refine": refine,
            "query": query
        }
        idx += 1
        new_data_pool.append(new_data)

print(len(new_data_pool))

output_path = "./../data/nq_search_extract_docs.jsonl"
write_jsonl(output_path, new_data_pool)
```
